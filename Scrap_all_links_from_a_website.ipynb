```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import urllib.parse
from urllib.parse import urlparse

# Function to scrape links recursively from the same domain
def scrape_links(url, visited_links, base_domain):
    # Skip if the URL is already visited
    if url in visited_links:
        return []

    # Mark the URL as visited
    visited_links.add(url)

    print(f"Scraping: {url}")
    all_links = []

    try:
        # Launch WebDriver and fetch the page
        driver.get(url)
        time.sleep(3)  # Wait for the page to load
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        # Find all <a> tags
        links = soup.find_all('a')
        for link in links:
            try:
                href = link['href']

                # Resolve relative URLs
                full_url = urllib.parse.urljoin(url, href)

                # Check if the URL belongs to the same domain
                parsed_url = urlparse(full_url)
                if parsed_url.netloc == base_domain:
                    all_links.append(full_url)
            except KeyError:
                continue

    except Exception as e:
        print(f"Error scraping {url}: {e}")

    # Recursively scrape links from the new pages
    for link in all_links:
        if link not in visited_links:
            all_links += scrape_links(link, visited_links, base_domain)

    return all_links


# Main code
if __name__ == '__main__':
    # Initialize WebDriver
    driver = webdriver.Chrome()
    driver.maximize_window()

    # Starting URL
    start_url = 'https://agecuttingtools.com/md8-601-carbide-tipped-general-purpose-8-inch-dia-x-60t-tcg-10-deg-5-8-bore.html'

    # Extract the base domain from the starting URL
    parsed_start_url = urlparse(start_url)
    base_domain = parsed_start_url.netloc

    # Set to track visited links
    visited_links = set()

    # Start scraping
    all_scraped_links = scrape_links(start_url, visited_links, base_domain)

    # Close WebDriver
    driver.quit()

    # Print all scraped links
    print(f"Total links scraped: {len(all_scraped_links)}")
    for link in all_scraped_links:
        print(link)
```
